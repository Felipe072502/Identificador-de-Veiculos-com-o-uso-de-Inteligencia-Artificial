fonte: https://www.gov.br/transportes/pt-br/assuntos/transito/conteudo-Senatran/frota-de-veiculos-2024


O Dataset inclui a quantidade de todos os veículos da Frota Nacional registrados até Agosto de 2024. A base foi tratada via powerquery do excel, onde foram categorizados as colunas em: UF, marca, modelo, ano de fabricação e quantidade de veículos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carrega os dados do arquivo Excel para um DataFrame
df = pd.read_excel('Frota Veículos Brasil Agosto 2024.xlsx')
display(df)
```

```python
# Exibe as últimas linhas do DataFrame
df.tail()
```

```python
# Mostra informações sobre o DataFrame, como tipos de dados e valores ausentes
df.info()
```

```python
# Gera estatísticas descritivas para colunas numéricas
df.describe()
```

```python
# Cria histogramas para cada coluna numérica
df.hist(figsize=(10, 8))
plt.show()
```

```python
# Cria gráficos de dispersão em pares para visualizar relacionamentos entre colunas
sns.pairplot(df)
plt.show()
```
